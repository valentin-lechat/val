```
#IMPORTATION DES LIBRAIRIES
import pandas as pd
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
import io
from google.colab import files
```

```
#IMPORT DES FICHIERS
uploaded = files.upload()
```

```
donnees = pd.read_csv(io.BytesIO(uploaded['04-28-2020.csv']))
donnees_US = donnees.loc[donnees['Country_Region']=="US",:]
donnees_UK = donnees.loc[donnees['Country_Region']=="United Kingdom",:]
donnees_ARGENTINE = donnees.loc[donnees['Country_Region']=="Argentina",:]
donnees_AUSTRALIE = donnees.loc[donnees['Country_Region']=="Australia",:]
donnees_BRESIL = donnees.loc[donnees['Country_Region']=="Brazil",:]
donnees_CANADA = donnees.loc[donnees['Country_Region']=="Canada",:]
donnees_CHINE = donnees.loc[donnees['Country_Region']=="China",:]
donnees_COLOMBIE = donnees.loc[donnees['Country_Region']=="Colombia",:]
donnees_FRANCE = donnees.loc[donnees['Country_Region']=="France",:]
donnees_RUSSIE = donnees.loc[donnees['Country_Region']=="Russia",:]
donnees_ALLEMAGNE = donnees.loc[donnees['Country_Region']=="Germany",:]
donnees_AUSTRALIE = donnees.loc[donnees['Country_Region']=="Australia",:]
donnees_JAPON = donnees.loc[donnees['Country_Region']=="Japan",:]
donnees_MEXIQUE = donnees.loc[donnees['Country_Region']=="Mexico",:]
donnees_SENEGAL = donnees.loc[donnees['Country_Region']=="Senegal",:]
donnees_AF_SUD = donnees.loc[donnees['Country_Region']=="South Africa",:]
donnees_ALGERIE = donnees.loc[donnees['Country_Region']=="Algeria",:]
Total_US=donnees_US['Confirmed'].sum()
Total_UK=donnees_UK['Confirmed'].sum()
Total_ARGENTINE=donnees_ARGENTINE['Confirmed'].sum()
Total_BRESIL=donnees_BRESIL['Confirmed'].sum()
Total_CANADA=donnees_CANADA['Confirmed'].sum()
Total_CHINE=donnees_CHINE['Confirmed'].sum()
Total_COLOMBIE=donnees_COLOMBIE['Confirmed'].sum()
Total_FRANCE=donnees_FRANCE['Confirmed'].sum()
Total_RUSSIE=donnees_RUSSIE['Confirmed'].sum()
Total_ALLEMAGNE=donnees_ALLEMAGNE['Confirmed'].sum()
Total_AUSTRALIE=donnees_AUSTRALIE['Confirmed'].sum()
Total_JAPON=donnees_JAPON['Confirmed'].sum()
Total_MEXIQUE=donnees_MEXIQUE['Confirmed'].sum()
Total_SENEGAL=donnees_SENEGAL['Confirmed'].sum()
Total_AF_SUD=donnees_AF_SUD['Confirmed'].sum()
Total_ALGERIE=donnees_ALGERIE['Confirmed'].sum()
Totaux=[Total_UK,Total_ARGENTINE,Total_BRESIL,Total_CANADA,Total_CHINE,Total_COLOMBIE,Total_FRANCE,Total_RUSSIE,Total_ALLEMAGNE,Total_AUSTRALIE,Total_JAPON,Total_MEXIQUE,Total_SENEGAL,Total_AF_SUD,Total_ALGERIE]
Pays=['UK','ARGENTINE','BRESIL','CANADA','CHINE','COLOMBIE','FRANCE','RUSSIE','ALLEMAGNE','AUSTRALIE','JAPON','MEXIQUE','SENEGAL','AF_SUD','ALGERIE']
y_pos=np.arange(len(Pays))
plt.bar(y_pos,Totaux,width=0.6,align='center',alpha=0.9)
plt.xticks(y_pos,Pays,rotation=70)
plt.ylabel('Nombre de cas confirmés (en Millions de personnes)')
plt.title('Nombre de cas confirmés de COVID 19 dans 16 Pays du Monde')
Total_confirmé_MONDE=donnees['Confirmed'].sum()
Total_décès_MONDE=donnees['Deaths'].sum()
Total_guéris_MONDE=donnees['Recovered'].sum()
Taux_de_mortalité_MONDE=(Total_décès_MONDE/Total_confirmé_MONDE)*100
Taux_de_personnes_touchées_mondialement=(Total_confirmé_MONDE/7781100000)*100
print("Nombre de cas confirmés dans le monde :",Total_confirmé_MONDE,"cas")
print("Nombre de décès dans le monde :",Total_décès_MONDE,"décès")
print("Nombre de guéris dans le monde :",Total_guéris_MONDE,"guéris")
print("Taux de mortalité mondial du virus :",Taux_de_mortalité_MONDE,"%")
print("Ratio des personnes touchées dans le monde par le virus :",'%.4f'%Taux_de_personnes_touchées_mondialement,"%")

```
